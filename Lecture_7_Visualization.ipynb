```python
# import libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
x = np.linspace(0,10,100)
fig, ax = plt.subplots(figsize = (6,4)) # tao figure va axes
ax.plot(x, np.sin(x))
ax.set_title('My asset')
ax.set_xlabel('X label')
plt.show()
```

```python
import matplotlib.pyplot as plt

# To Figure v Axes
fig, ax = plt.subplots(figsize=(7, 5))

# V scatter theo tng nhm time
for t in tips["time"].unique():
    subset = tips[tips["time"] == t]
    ax.scatter(
        subset["total_bill"],
        subset["tip"],
        label=t,
        alpha=0.7,
        s=60
    )

# Thit lp nhn v tiu 
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
ax.set_title("Tip vs Total Bill (colored by Time)")

# Legend
ax.legend(title="Time")

plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, 
                         ncols=2, 
                         figsize = (4,6),
                         sharex= True,
                         sharey= False) 
axes[0,0].plot(x,np.sin(x))
axes[0,0].set_title('$Sin(x)$ function')
axes[1,1].plot(x,np.cos(x));
```

```python
layout = [['l','c','r'],
          ['l1','c','r'],
          ['l3',"l3",'r']]
fig, axes = plt.subplot_mosaic(layout)
axes['r'].plot(np.sin(x),x)
axes['l1'].plot(x, np.cos(x));
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Ly d liu tip
tip_data = tips["tip"].dropna()

# To Figure & Axes
fig, ax = plt.subplots(figsize=(7, 5))

# Histogram (chun ho  v KDE)
ax.hist(
    tip_data,
    bins=20,
    density=True,
    alpha=0.6,
    edgecolor="black",
    label="Histogram"
)

# KDE
kde = gaussian_kde(tip_data)
x = np.linspace(tip_data.min(), tip_data.max(), 300)
ax.plot(x, kde(x), linewidth=2, label="KDE")

# Nhn & tiu 
ax.set_xlabel("Tip amount")
ax.set_ylabel("Density")
ax.set_title("Distribution of Tip with KDE")

ax.legend()
plt.show()
```
