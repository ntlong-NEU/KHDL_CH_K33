## Lecture 4: IF-ELSE and While loops

```python
age = int(input('How old are you?'))
discount = 0
if age<10:
    discount = 0.15
elif age<18:
    discount = 0.08
elif age<23:
    discount = 0.05
elif age<=60:
    discount = 0.02
else: 
    discount = 0.25
p = 200
print(f'Đơn hàng được giảm {discount*100} %')
print(f'Tổng đơn hàng là: {p}, Số tiền cần thanh toán sau giảm giá là: {p*(1-discount)}')

```

```python
x = 1
while True:
    print(x)
    if x>4:
        break
    x +=1 
```

```python
x = 1
while x<=4:
    print(x)
    x +=1 
```

```python
x = 1
active = True
while active:
    print(x)
    if x>4:
        active = False
    x +=1 
```
