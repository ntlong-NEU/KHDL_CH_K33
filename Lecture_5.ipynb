```python
def normlize_text(text:str, lower = True):
    """docstring"""
    text_norm = text
    if lower:
        text_norm = text_norm.lower()
    special_char = {'.',',','-','!',':','*','?'}
    for char in special_char:
        text_norm = text_norm.replace(char, ' ')
    return text_norm
    
```
